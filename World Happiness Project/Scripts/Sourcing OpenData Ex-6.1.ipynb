```python
def check_mixed_dtypes(df):
    mixed_columns = {}
    for column in df.columns:
        unique_dtypes = df[column].map(type).unique()
        if len(unique_dtypes) > 1:
            mixed_columns[column] = unique_dtypes
    return mixed_columns
```

```python
mixed_dtypes = check_mixed_dtypes(df)
```

```python
print("Columns with mixed data types:")
for column, dtypes in mixed_dtypes.items():
    print(f"{column}: {dtypes}")
```
